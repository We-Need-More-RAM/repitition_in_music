# How to make a Snowflake Diagram with Pandas and Seaborn

## Imports

```python
import re

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

## Process

### We will use a simple string as a test

```python
sample_string = 'Mary had a little lamb, whose fleece was white as snow. And everywhere that Mary went, the lamb was sure to go.'
sample_string
```

### First we will remove everything but alphanumeric characters and spaces

```python
formatted_sample_string = re.sub(r'[^A-Za-z0-9 ]+', '', sample_string)
formatted_sample_string
```

### Then we will create a list that holds each individual word

```python
formatted_sample_string_words = formatted_sample_string.split()
formatted_sample_string_words
```

### Now we will create a dataframe where the column and index names are each word of the lyrics

```python
df = pd.DataFrame([], columns=formatted_sample_string_words, index=formatted_sample_string_words).fillna(0)
df
```

### And then we will assign a 1 to every cell where the column and index name match

```python
intersection = df.index & df.columns

for word in intersection:
    df.loc[word, word] = 1

df
```

### Now we can create a heatmap of the dataframe to see the intersections

```python
plt.figure(figsize = (10,10))
sns.heatmap(df, cmap='Blues')
```

### And now let's dress up that graph and remove unimportant info

```python
fig, ax = plt.subplots(figsize = (10,10))
graph = sns.heatmap(df, cmap='Blues', cbar=False, xticklabels=False, yticklabels=False)

ax.set_title("Mary had a little lamb", 
fontsize=30, fontweight="bold", pad=20)
fig.set_facecolor("white")
ax.set_facecolor('white')
plt.show()
```
